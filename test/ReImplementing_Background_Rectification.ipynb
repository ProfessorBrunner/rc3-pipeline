Finally finished debugging the whole thing and it works now. The reason why there was all these mistakes and bugs was that I was not careful with what goes into and out of a command. In the example here http://montage.ipac.caltech.edu/docs/m101tutorial.html in steps 5 and 6 they have images.tbl, but in fact I should be putting in pimages.tbl ( also dont forget the plural s). I need to use pimages.tbl because these are the files that "Image metadata table describing the reprojected images", always read the file description on the right of the table (not just the filename because image.tbl is misleading)


Can also try exploring the textfile inside image tables and you will find that hdu0 only exist in projected images. Also make sure your input directory is "projected" when you are in stage 5 and your input directory is "corrdir". Another debugging tip is always look at whether a directory is empty or a table file is empty, then the problem stems from there. Montage is really bad at this kind of stuff.

Also becareful of cascading file structure (i.e. u contain g contain r contain i contain z)
