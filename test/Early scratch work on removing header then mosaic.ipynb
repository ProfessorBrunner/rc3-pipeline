```python
def pixel_hist(image_file):
    fitsImage = fits.open(image_file)
    import matplotlib.pyplot as plt
    image = fitsImage[0].data
    num_pixels = image.shape[0]*image.shape[1]
    print num_pixels, " pixels in image"
    fig = plt.figure()
    ax = fig.add_subplot(111)
    pixels = list(flatten(image)) #normalized pixels 
    max_pix =max(pixels)
    m = mean(pixels)
    s = std(pixels)
    print "Mean: ",m
    print "Std: ",s
    weights = np.ones_like(pixels)/len(pixels)
    plt.hist(pixels,bins=1000, weights=weights)
    plt.xlim(0,m+3*s)
    plt.xlabel("Pixel Value")
    plt.ylabel("Pixel Count")
    plt.title("Pixel Histogram Plot")
```
