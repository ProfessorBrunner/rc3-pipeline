```python
import mpld3
# mpld3.enable_notebook()
mpld3.disable_notebook()
```

```python
gal = np.where(input[::,3]>15.)
gal2 = np.intersect1d(np.where(input[::,3]<15.)[0], np.where(input[::,3]>10.)[0])
gal3 = np.intersect1d(np.where(input[::,3]<10.)[0], np.where(input[::,3]>5.)[0])
gal4 = np.where(input[::,3]<5.)[0]
pltx.fit_and_plot(input[::,2],output[::,2],xlabel="Input Magnitude",ylabel="Output Magnitude",title="500 RC3 Random Test",marker=',')
plt.plot(input[::,2][gal],output[::,2][gal],'.',color = 'cyan', label = "Galaxies (Sources r>15pix)")
plt.plot(input[::,2][gal2],output[::,2][gal2],'.',color = 'lime', label = "Stars (Sources 15>r>10pix)")
plt.plot(input[::,2][gal3],output[::,2][gal3],'.',color = 'fuchsia', label = "Stars (Sources 10>r>5pix)")
plt.plot(input[::,2][gal4],output[::,2][gal4],'.',color = 'blue', label = "Stars (Sources r<5pix)",markersize=2)
plt.plot(input[::,2][846],output[::,2][846],'x',color = 'red', label = "Outlier PGC 7328",markersize=10)
plt.plot(input[::,2][331],output[::,2][331],'x',color = 'blue', label = "Outlier PGC 24286",markersize=10)
plt.plot(input[::,2][690],output[::,2][690],'x',color = 'green', label = "Outlier PGC 1089",markersize=10)
plt.plot(input[::,2][498],output[::,2][498],'x',color = 'magenta', label = "Outlier PGC 57117",markersize=10)
plt.plot(input[::,2][236],output[::,2][236],'x',color = 'orange', label = "Outlier PGC 57117",markersize=10)
plt.plot(input[::,2][1500],output[::,2][1500],'x',color = 'turquoise', label = "Outlier PGC 57117",markersize=10)


from matplotlib.legend_handler import HandlerLine2D
# plt.legend(loc='upper left',prop={'size':8},numpoints=1)

pltx.fit_and_plot(input[::,2],output[::,2]-input[::,2],xlabel="Input Magnitude",ylabel="(Output-Input) Magnitude",title="500 RC3 Random Test",marker=',')
# plt.plot(input[::,2][601],(output[::,2]-input[::,2])[601],'x',color = 'red')
plt.plot(input[::,2][gal],(output[::,2]-input[::,2])[gal],'.',color = 'cyan', label = "Galaxies (Sources r>15pix)")
plt.plot(input[::,2][gal2],(output[::,2]-input[::,2])[gal2],'.',color = 'lime', label = "Stars (Sources 15>r>10pix)")
plt.plot(input[::,2][gal3],(output[::,2]-input[::,2])[gal3],'.',color = 'fuchsia', label = "Stars (Sources 10>r>5pix)")
plt.plot(input[::,2][gal4],(output[::,2]-input[::,2])[gal4],'.',color = 'blue', label = "Stars (Sources r<5pix)",markersize=2)
plt.plot(input[::,2][846],(output[::,2]-input[::,2])[846],'x',color = 'red', label = "Outlier PGC 7328",markersize=10)
plt.plot(input[::,2][331],(output[::,2]-input[::,2])[331],'x',color = 'blue', label = "Outlier PGC 24286",markersize=10)
plt.plot(input[::,2][690],(output[::,2]-input[::,2])[690],'x',color = 'green', label = "Outlier PGC 1089",markersize=10)
plt.plot(input[::,2][498],(output[::,2]-input[::,2])[498],'x',color = 'magenta', label = "Outlier PGC 57117",markersize=10)
plt.plot(input[::,2][236],(output[::,2]-input[::,2])[236],'x',color = 'orange', label = "Outlier PGC 57117",markersize=10)
plt.plot(input[::,2][1500],(output[::,2]-input[::,2])[1500],'x',color = 'turquoise', label = "Outlier PGC 57117",markersize=10)
# plt.legend(loc='lower left',prop={'size':12},numpoints=1)
```
