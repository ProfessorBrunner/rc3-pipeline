```python
input = np.loadtxt("input_mag")
output = np.loadtxt("output_mag")
```

```python
import pltx
```

```python
import mpld3
mpld3.enable_notebook()
mpld3.disable_notebook()
```

```python
outlier_n = np.intersect1d(np.where(input[::,2]>-7.5995)[0],np.where(input[::,2]<-7.5990)[0])[0]
```

```python
outlier_n
```

```python
pltx.fit_and_plot(input[::,2],output[::,2],xlabel="Input Magnitude",ylabel="Output Magnitude",title="500 RC3 Random Test",marker=',')
plt.plot(input[::,2][outlier_n],output[::,2][outlier_n],'x',color = 'red', label = "Outlier",markersize=10)
plt.legend(loc='lower right',prop={'size':12},numpoints=1)

pltx.fit_and_plot(input[::,2],output[::,2]-input[::,2],xlabel="Input Magnitude",ylabel="(Output-Input) Magnitude",title="500 RC3 Random Test",marker=',')
plt.plot(input[::,2][outlier_n],(output[::,2]-input[::,2])[outlier_n],'x',color = 'red', label = "Outlier",markersize=10)
plt.legend(loc='lower right',prop={'size':12},numpoints=1)
```
