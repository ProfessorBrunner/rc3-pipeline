```python
input = np.loadtxt("input_mag")
output = np.loadtxt("output_mag")
```

```python
import pltx
pltx.fit_and_plot(input[::,2],output[::,2],xlabel="Input Magnitude",ylabel="Output Magnitude",title="500 RC3 Random Test",marker=',')
plt.plot(input[::,2][601],output[::,2][601],'x',color = 'red')
#Delta Mag
pltx.fit_and_plot(input[::,2],output[::,2]-input[::,2],xlabel="Input Magnitude",ylabel="(Output-Input) Magnitude",title="500 RC3 Random Test",marker=',')
plt.plot(input[::,2][601],(output[::,2]-input[::,2])[601],'x',color = 'red')
```

SExtractor records flux_radius in units of pixels , this is stored inside our ``input_mag``

    FLUX_RADIUS            Fraction-of-light radii                                    [pixel]
Inside rc3.py (line 368) , we enforce that a "detected rc3 source" is anything with radius of 15 pixel or above, so we use the same criterium here for star galaxy separation. 


```python
gal = np.where(input[::,3]>15.)
```

```python
pltx.fit_and_plot(input[::,2],output[::,2],xlabel="Input Magnitude",ylabel="Output Magnitude",title="500 RC3 Random Test",marker=',')
plt.plot(input[::,2][gal],output[::,2][gal],'o',color = 'red', label = "Galaxies (Sources r>15pix)")
from matplotlib.legend_handler import HandlerLine2D
plt.legend(loc='lower right',prop={'size':12},numpoints=1)

pltx.fit_and_plot(input[::,2],output[::,2]-input[::,2],xlabel="Input Magnitude",ylabel="(Output-Input) Magnitude",title="500 RC3 Random Test",marker=',')
# plt.plot(input[::,2][601],(output[::,2]-input[::,2])[601],'x',color = 'red')
plt.plot(input[::,2][gal],(output[::,2]-input[::,2])[gal],'o',color = 'red', label = "Galaxies (Sources r>15pix)")
plt.legend(loc='lower right',prop={'size':12},numpoints=1)
```
