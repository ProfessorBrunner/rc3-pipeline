```python
mega_ra = []
mega_dec = []
coord =[]
with open("rc3_ra_dec_diameter_pgc.txt",'r') as f:
    for line in f:
        info = line.split()
        mega_ra.append(float(info[0]))
        mega_dec.append(float(info[1]))
        coord.append([float(info[0]),float(info[1]),float(info[2]),int(info[3])])
coord=np.array(coord)
import random
rc3_subsample = []
for i in np.arange(500):
    n = randint(23011)
    rc3_subsample.append((coord[n,0],coord[n,1],coord[n,2],coord[n,3]))
rc3_subsample = np.array(rc3_subsample)
pltx.plot_mwd(rc3_subsample[:,0],rc3_subsample[:,1], org=90, title ='All RC3 Galaxies')
subsample_radius = rc3_subsample[::,2]
plt.figure()
histo = plt.hist(subsample_radius,100)
mu = mean(subsample_radius)
sig = np.std(subsample_radius)
plt.xlim(0,mu+3*sig)
plt.title("Size distribution of subsample RC3 Galaxies",fontsize=12)
plt.xlabel("Radius [deg]",fontsize=12)
plt.ylabel("Number of RC3 in subsample",fontsize=12)
np.savetxt("rc3_subsample.txt",rc3_subsample)
```
