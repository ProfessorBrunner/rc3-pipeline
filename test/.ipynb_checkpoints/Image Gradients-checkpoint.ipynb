```python
from matplotlib.colors import LogNorm
image = hdulist[0].data
plt.imshow(image,norm=LogNorm(),cmap=cm.Greys_r)
plt.show()
output=np.zeros_like(image)
# print output
for x in np.arange(1,len(image)-1): # Remember that arange doesn't include the ending element anyways
    output[:,x] = np.abs(image[:,x]-image [:,x+1])
plt.imshow(output,norm=LogNorm(),cmap=cm.Greys_r)
plt.show()
```

```python
from matplotlib.colors import LogNorm
image = hdulist[0].data
plt.imshow(image,norm=LogNorm(),cmap=cm.Greys_r)
plt.show()
output=np.zeros_like(image)
# print output
for y in np.arange(1,len(image)-1): # Remember that arange doesn't include the ending element anyways
    output[y,:] = np.abs(image[y,:]-image [y+1,:])
plt.imshow(output,norm=LogNorm(),cmap=cm.Greys_r)
plt.show()
```
