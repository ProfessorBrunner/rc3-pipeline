# _Server_

Data acquisition is composed of two main tasks: querying imaging data and retrieving data from server. Most surveys have an API that enables data access by SQL or customized query. To make the __Server__ class work, we need to figure out how to convert positional values to record-keeping parameters dependent on the survey's telescope. For example, SDSS image frames are uniquely identified by a particular combination of  run, camcol, and field and 2MASS refers to this with its sexagesimal, equatorial position-based source name. In the case where the survey has no special name referring to the imaging frame, such as in the case of POSS-II data, the user can invent their own naming scheme as placeholders for naming the raw imaging data. In addition to these two essential functions, each __Server__ class should also implement the query builder methods that are used inside the main mosaicking program.  Our choice of using a server instead of a data object enables code reusability across various surveys that uses common server tools , such as IPAC's GATOR query service or  Astroquery.

__Subclass__: Gator, SkyServer, MAST...etc

###### Abstract methods

- __otherRC3(ra,dec,margin)__:
 - Given ra,dec, pgc of an RC3 galaxy, use  Vizier to find a list of other rc3 that lies in the same margin field.
        in the form including the original galaxy of interest

        [['PGC54', 0.158083, 28.384556], ['PGC58', 0.183333, 28.401444]]

        Units
        =====
        Search radius (radius): arcsecond
        Search box (size): arcsecond
 - This method can be overriden if the survey contains its own query system for RC3 objects such as in SkyServer.
- __abstract  query (query): __:
interacts with the query system of the data server and returns a list of desirable information.
       args:
         string query command
         string mirror_database (Optional)
         int format (Optional):
         --> 1 :['#Table1\n', 'run,camcol,field\n', '1024,2,115\n', '1045,2,129\n']
         --> 2 :[['1024', '2', '115'], ['1045', '2', '129']]

       out: 
           List of matching result, depend on format, by format=1 default.
 
- __abstract getData(self,*args)__: 
 - Get data(*kwargs): interacts with the data server to retrieve Data by giving Surveys specific imaging names, for example run, camcol, field; ra,dec.

- __abstract surveyFieldConverter(ra,dec,clean,*args)__:
   - Returns which field the object lies in the survey
        the output and implementation of this method is hihgly dependent on survey structure
        Note: even though "clean" is optional, the parameter field stil needs to be written in to conform to method calls in rc3.py
        
        ex) SDSS => run,camcol,field
            2MASS => Tiles designation

# _Survey_

Survey is an abstract class that is implemented by the specific survey, For example SDSS, 2Mass, GALEX, mixed bands. Mixed bands may contain multiple surveys which enables it to mosaic from multiple surveys per band. This is particularly useful for GALEX because it only has two bands so it could be augmented by having a third then from one of the other surveys. These classes contain one of the concrete classes that implement server. 

Survey class must also contain the parameters and information used in the mosaic instead. The survey class is designed such that  It has the flexibility for other survey dependent applications  to latch on the survey class this independently, (VO table, navigator tools, API,NED?). 

__Subclass__ : SDSS, DSS, 2MASS, ...etc

Instance Attributes:

- General Info used for Mosaicing
    - str __name__
    - list __bands__ : ['u','g','r','i','z']
    - list __color_bands__: ['g','r','i']
    - char __best_band__ : best quality band, the band that best show low surface brightness structure (used for the zooming out algorithm)  'r' for SDSS, usually selected as the band with highest quantum efficiency.
    - float __pixel_scale__: for SExtractor in units of [arcsec/pix]
    - Server __data_server__: stores the Server object
    - str __sextractor_params__
    - str __montage_params__
    - str __stiff_params__

Methods
- _initServer(): Private method that type dispatches a specific server upon initialization

# _Catalog_

The abstract __Catalog__ class is a concrete class that is responsible for all of the automated mosaic steps(i.e. Mosaic all galaxies in the Given catalog). This generic catalog class also enables the flexibility of mosaic and other catalogs such as the UGC, PGC, and Messier  which would be subclasses of catalog class. The Catalog class contains CatalogObjects such as RC3Objects to make use of the per-object mosaicking methods.

__Subclass__ : RC3Catalog

Instance Attributes:

- __name__
- __frame__ : default 'fk5'
- __equinox__ : default 'J2000'
- __allObj__: a list of all objects obtained after calling _initAllObj()


Methods:
- __\_initAllObj()__: Private methods that creates a list of all objects inside the catalog
- __initSubset(file)__:Create a list of all objects inside a user-defined catalog
- __mosaicAll(survey)__:survey is a  Survey object. Produce all band FITS files and color mosaic for every objects inside the Catalog that lies within the footprint of the given survey

# RC3

The RC3 class represents an RC3 galaxy object. It is responsible for mosaic on a per object basis. It also contains information about the RC3 object including links to NED database and other information.

Instance Attributes:
  - __rc3_ra, rc3_dec,rc3_radius__: as stored in the original RC3 Catalog
  - __pgc__ : Unique identifier 
  - __num_iterations__: number of times source_info is executed.
  - __new_ra, new_dec__: Initialized as @; later updated by the positional update algorithm.
  - __new_radius__ :Updated Radius is not too reliable (dependent on definition, we return the isophotal radius that includes 50% of the light)

Methods:

- mosaic_band(self,band,ra,dec,margin,radius,pgc,survey):Create a mosaic fit file for the specified band.
 - Input: source info param
 - Return: String filename of resulting mosaic
        
- source_info(self,r_fits_filename,survey):
 - Input: Filename String of R band Mosaic fit file
 - Returns the updated [ra,dec,margin,radius,pgc] info about the identified RC3 source as a list
    - If no RC3 source is identified then ['@','@',margin_value,'@','@'] is returned
    - If RC3 lie outside of survey footprint then [-1,-1,-1,-1,-1] is returned
    
- __mosaic_all_bands(ra,dec,margin,radius,pgc,survey)__:Creates u,g,r,i,z fit file mosaic and g,r,i color images.Call this on the final step of the program,once we have verified that the RC3 source lies inside the new margin and updated center ra,dec and radii.
